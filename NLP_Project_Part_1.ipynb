## **Part 2 i. GPT-2 117M**

```
gpt2.download_gpt2(model_name="117M")
```

**Mounting Google Drive**

```
gpt2.mount_gdrive()
```

**Upload Sample PS to colab**

```
file_name = '/content/drive/My Drive/NLP/Triage-TokenSize__Core.ps1'
```

```
import tensorflow as tf
tf.reset_default_graph()
```

```
sess = gpt2.start_tf_sess()

gpt2.finetune(sess,
              dataset=file_name,
              model_name='117M',
              steps=1000,
              restore_from='fresh',
              print_every=10,
              sample_every=200,
              save_every=500,
              )
```

```
ls checkpoint/run1
```

**Copy the checkpoint folder to own Google Drive**

```
gpt2.copy_checkpoint_to_gdrive()
```

**Load a Trained Model Checkpoint**

```
gpt2.copy_checkpoint_from_gdrive()
```

Load the retrained model checkpoint + metadata necessary to generate text.

**IMPORTANT NOTE:** If you want to rerun this cell, restart the VM first (Runtime -> Restart Runtime). You will need to rerun imports but not recopy files.

```
sess = gpt2.start_tf_sess()
gpt2.load_gpt2(sess)
```

**Generate Text From The Trained Model**

```
gpt2.generate(sess)
```

Let's generate more samples with a specific prefix `$Server = `

In actual script, this is a declaration of Server IP address. 

```
gpt2.generate(sess,
              length=1023,
              temperature=0.7,
              prefix="$Server =",
              nsamples=5,
              batch_size=5
              )
```
